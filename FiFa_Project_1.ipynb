```python
def convert_suffix_to_float(i):
    suffix = i[-1].upper()
    if suffix == 'M':
        return float(i[:-1]) * 1_000_000
    elif suffix == 'K':
        return float(i[:-1]) * 1_000
    else:
        return float(i)
```
